```python
%matplotlib inline
import ipython_genutils
import numpy as np

import  matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import torch
import torch.nn as nn

plt.ion()
```

# Reggression losses

Options
- L1 loss
- MSE loss

```python
losses = {'L1 loss' : nn.L1Loss(reduction='none'), 'MSE loss' : nn.MSELoss(reduction='none')}

for i, loss in enumerate(losses):

    plt.subplot(len(losses), 2, 2 * i + 1)
    plt.title(loss)

    output = torch.zeros(100)
    labels = torch.arange(0,100)

    loss_val = losses[loss](output, labels)

    plt.plot(labels, loss_val)

plt.show()
```
