```python
import numpy as np

import gym

import time
```

```python
env = gym.make("FrozenLake-v0")
print(env.observation_space)
```

```python
env.reset()
while True:
    action = env.action_space.sample()
    obs, rew, done, info = env.step(action)
    env.render()

    if(done):
        print("Dead")
        break

    time.sleep(0.5)
```
